```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Orginal

```python

image_path = "./data/me.jpg"
image = cv2.imread(image_path)

plt.imshow(image, cmap='gray')
plt.title('Original Image')
```

## Crop Image

```python
# set mask
start_row, start_col = 15, 15

#create Mask
mask = np.zeros((3, 3, 3), dtype=np.uint8)
mask[:, :] = image[start_row:start_row + 3, start_col:start_col + 3]
print(mask[0:3,0:3])
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))  # OpenCV خوانده شده به RGB تبدیل می‌شود
plt.title('Mask (3x3)')
plt.axis('off')  # غیرفعال کردن محورهای تصویر
plt.show()
```

```python
kernel = mask[:,:,1]

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Crop Image Convolution Result')

plt.show()
```

## Sharpen

```python

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Sharpen Convolution Result')

plt.show()
```

## Laplacian

```python

kernel = np.array([[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Laplacian Convolution Result')

plt.show()
```

## Emboss

```python

kernel = np.array([[-2, -1, 0],
                   [-1, 1, 1],
                   [0, 1, 2]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Emboss Convolution Result')

plt.show()
```

## Outline

```python

kernel = np.array([[-1,-1,-1],
                   [-1, 8, -1],
                   [-1,-1,-1]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Outline Convolution Result')

plt.show()
```

## Bottom Sobel

```python

kernel = np.array([[-1,-2,-1],
                   [0,0,0],
                   [1,2,1]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Bottom Sobel Convolution Result')

plt.show()
```

## Right Sobol

```python

kernel = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Right Sobol Convolution Result')

plt.show()
```

## Top Sobol

```python

kernel = np.array([[1,2,1],
                   [0,0,0],
                   [-1,-2,-1]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Top Sobol Convolution Result')

plt.show()
```

## Weighted Average

```python

kernel = np.array([[1,1,1],
                   [1,8,1],
                   [1,1,1]])

result = cv2.filter2D(image, -1, kernel)

plt.imshow(result, cmap='gray')
plt.title('Weight Average Convolution Result')

plt.show()
```
